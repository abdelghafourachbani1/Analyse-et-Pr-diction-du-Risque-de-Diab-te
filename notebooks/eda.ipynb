Import Data

```python
import pandas as pd

df = pd.read_csv('../data/raw/dataset-diabete.csv',index_col=0)

print(df.shape) 
print(df.info())
print(df.describe())
print(df.isnull().sum())
print(df.duplicated().sum())
```

Histogram & hotmap

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
df.hist(figsize=(14, 10), bins=20, color='#00adb5', edgecolor='black')
plt.suptitle("Distribution of Numerical Variables", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Variables", fontsize=16)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('default')

plt.figure(figsize=(10, 3))
# Draw the boîte à moustaches
sns.boxplot(x=df['Age'], color='#00adb5')
plt.title("Boîte à Moustaches - Pregnancies", fontsize=14)
plt.show()
```
